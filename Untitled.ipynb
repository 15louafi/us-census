This US Census dataset contains detailed but anonymized information for approximately 300,000 people.

The archive contains 3 files:
o A large learning .csv file
o Another test .csv file
o A metadata file describing the columns of the two above mentioned files (identical for both)

The goal of this exercise is to model the information contained in the last column (42nd), i.e., which people make more or less than $50,000 / year, from the information contained in the other columns. The exercise here consists of modeling a binary variable. 

# **Part 1: Obtaining the Data**

```python
#Importing some useful libraries
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
%matplotlib inline
```

```python
#Opening the dataset and saving it on df
df = pd.read_csv("census_income_learn.csv",header=None, na_values=[" ?", "?"])
df_test = pd.read_csv("census_income_test.csv",header=None, na_values=[" ?", "?"])
```

```python
#Let's get a quick overview of our dataset
df.head(5)
```

At this point, we have no column names.

```python
#Let's get the column headers from the metadata file for better readability
f = open('census_income_metadata.txt', 'r') #open file for reading

def build_data(f): #build our list

    output = [] #List
    for line in f.readlines(): #Iterates through every line in the file
        if ":" in line: #make sure line has data as wanted
            l = line.split(":") #Seperate by ":" and put into a list
            if "|" not in l[0].strip(): #Removes the tabs (so we only keep the column names at the end of the file)
                output.append(l[0].strip()) #Then get rid of the the starting and end character and append them
    output.append("income") #Add the income column
    return output #Returns the list
data = build_data(f)

print(data) #Checking if we got what we wanted
print(len(data)) #And if we have not forgotten any column name
```

```python
#Now let's add the column names to our dataframe !
df.columns = data
df_test.columns = data
print(df.head(5))
```

Now we're ready to comb through our dataset !

# Part 3 : Feature Exploration

Let's deal with the missing values right away. One strategy would be to not use the columns into account.

For the migration codes, seems like almost 50% of the data is missing. Judging by the metadata, it's pretty hard to manually predict the values, and doing so would add a lot of noise into our data (since we have 50% of the values missing). I'm going to rid of these of these columns: although they do contain some information, trying to fill the missing values is not worth it.



```python
df = df.drop(['migration code-change in msa',
'migration code-change in reg'
,'migration code-move within reg'
,'migration prev res in sunbelt'], axis=1)
df_test = df_test.drop(['migration code-change in msa',
'migration code-change in reg'
,'migration code-move within reg'
,'migration prev res in sunbelt'], axis=1)
```

For country of birth father, country of birth mother and country of birth self, if the data is missing, that probably means that the country is not on the metadata list (since the country list is not exhaustive). No need to do anything then, when we will perform one-hot encoding, then this phenomena will be taken into account. Same thing applies for state of previous residence.

Just with my instinct (and some common sense), I believe some of the variables that are more likely to be correlated with the income are :
- wage per hour (if it's high then your income is high)
- age (the older you are, the higher your wage)

- education

# Part 4 : Feature Engineering

Based on some of the columns that we've got, we can roughly estimate the income, using wage_per_hour, weeks worked in year, dividends from stocks, capital gains and capital losses, cosnidering that the average US citizen works 40 hours a week.

```python
df_ohe["new_feature"]=(40 * df_ohe["wage per hour"] * df_ohe["weeks worked in year"]) + df_ohe["dividends from stocks"] + df_ohe["capital gains"] + df_ohe["capital losses"]
df_test_ohe["new_feature"]=(40 * df_test_ohe["wage per hour"] * df_test_ohe["weeks worked in year"]) + df_test_ohe["dividends from stocks"] + df_test_ohe["capital gains"] + df_test_ohe["capital losses"]
```

Now let's use variance threshold to get rid of some useless variables.

```python
print(df_ohe.head())
```

```python
from sklearn.feature_selection import VarianceThreshold
# Remove features with low variance, thus which explain the data poorly
vt = VarianceThreshold(threshold=(0.05)) #We set 0.05 as our variance threshold
vt.fit(df_ohe)
indices = vt.get_support(indices=True) #We get the indices of the top features
print(indices)
X_train = df_ohe.iloc[:, indices] #We extract the new data
Y_train = X_train["income_ 50000+."] #BuildY vector
X_train.drop("income_ 50000+.", axis=1, inplace=True)
X_test = df_test_ohe[list(X_train)]
Y_test = df_test_ohe["income_ 50000+."]
X_train.describe(include='all')

```

We can push the feature engineering further by grouping certain variables in buckets or creating new features, but we can stick with that for now.

# Part 5 : Random Forest and Feature Selection

Now that we've got our preprocessing/feature engineering done, we can actually start running our first classifier! My choice will be Random Forest:
- It works with categorical data, so we can run it even without one-hot encoding.
- We can get the feature importances and thus allow us to perform an even better feature selection (and find which variables influence the income the most).
- It works well even with outliers
- It fits well with high dimensional data.

```python
from sklearn.cross_validation import cross_val_score
from sklearn.feature_selection import SelectPercentile, SelectKBest, f_classif, chi2
from sklearn import metrics
import numpy as np
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, ExtraTreesClassifier, AdaBoostClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.discriminant_analysis import QuadraticDiscriminantAnalysis
names = ["Logistic Regression", 
         "Nearest Neighbors", 
#          "Linear SVM", "RBF SVM", 
         "Decision Tree",
         "Random Forest", "Extra Trees", "AdaBoost", "Naive Bayes", 
         "Linear Discriminant Analysis",
         "Quadratic Discriminant Analysis"]

logReg = LogisticRegression()
knc = KNeighborsClassifier(3)
# hard to scale to dataset with more than a couple of 10000 samples:
# linsvm = SVC(kernel="linear", C=0.025)
# rbfsvm = SVC(gamma=2, C=1)
dectree = DecisionTreeClassifier()
randforest = RandomForestClassifier(n_estimators=500, min_samples_leaf=5, n_jobs=-1)
exTree = ExtraTreesClassifier()
adaBoost = AdaBoostClassifier()
naiveBayes = GaussianNB()
lda = LinearDiscriminantAnalysis()
quadlda = QuadraticDiscriminantAnalysis()

allmodels = [
    logReg,
    knc,
#     linsvm,
#     rbfsvm,
    dectree,
    randforest,
    exTree,
    adaBoost,
    naiveBayes,
    lda,
    quadlda]

scores = []
for name, model in zip(names, allmodels):
    model.fit(X_train, Y_train)
    score = model.score(X_test, Y_test,scoring='roc_auc')
    scores.append((model, name, score))
    print('{} : {}'.format(name, score))
```
